# Statistics Part 2

1. What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves making an assumption (the null hypothesis) and then using sample data to assess whether to reject or fail to reject that assumption.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - Null Hypothesis (H₀): A default statement that there is no effect, difference, or relationship.

  - Alternative Hypothesis (H₁ or Ha): The statement that contradicts the null hypothesis and represents what the researcher wants to prove.

  - Example:

    - H₀: μ = 50 (no change)

    - H₁: μ ≠ 50 (there is a change)



3. What is the significance level in hypothesis testing, and why is it important?
  - Significance Level (α): The threshold probability for rejecting the null hypothesis. Common values: 0.05, 0.01.

  - It defines the maximum acceptable probability of making a Type I error (false positive).

  - Lower α → more stringent test.

4. What does a P-value represent in hypothesis testing?
  - The p-value is the probability of observing the sample result or more extreme, assuming the null hypothesis is true.

  - It quantifies the evidence against the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?
  - If p-value ≤ α: Reject the null hypothesis (significant result).

  - If p-value > α: Fail to reject the null hypothesis (not significant).

6. What are Type 1 and Type 2 errors in hypothesis testing?
  - Type I Error (False Positive): Rejecting H₀ when it is true. Probability = α.

  - Type II Error (False Negative): Failing to reject H₀ when H₁ is true. Probability = β.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - One-tailed test: Tests for a deviation in only one direction (e.g., H₁: μ > μ₀).

  - Two-tailed test: Tests for deviations in both directions (e.g., H₁: μ ≠ μ₀).

8. What is the Z-test, and when is it used in hypothesis testing?
  - A Z-test is used when:

    - Population variance is known.

    - Sample size is large (n > 30).

  - Used to compare sample mean to a known population mean.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - Z = (X̄ - μ) / (σ / √n)

    - X̄ = sample mean, μ = population mean, σ = population std. deviation, n = sample size.

  - The Z-score tells how many standard deviations the sample mean is from the population mean.



10. What is the T-distribution, and when should it be used instead of the normal distribution?
  - Used when:

    - Sample size is small (n < 30).

    - Population standard deviation is unknown.

  - T-distribution is wider and has heavier tails than the normal distribution.

11. What is the difference between a Z-test and a T-test?
  - The Z-test is used when the population standard deviation is known and the sample size is large (usually > 30); it is based on the standard normal distribution and is ideal for testing population means or proportions with known variance.

  - The T-test is used when the population standard deviation is unknown and the sample size is small (≤ 30); it relies on the t-distribution, which has heavier tails to account for greater uncertainty, and becomes similar to the Z-test as the sample size increases.

12. What is the T-test, and how is it used in hypothesis testing?
  - T-test is used to compare means when sample size is small, and variance is unknown.

  - Types:

    - One-sample t-test: Compare sample mean with population mean.

    - Two-sample t-test: Compare two independent sample means.

    - Paired t-test: Compare means of related samples.

13. What is the relationship between Z-test and T-test in hypothesis testing?
  - Both are used to compare means.

  - T-test is a generalization of the Z-test for small samples.

  - As sample size increases, the t-distribution approximates the normal (Z) distribution.

14. What is a confidence interval, and how is it used to interpret statistical results?
  - A confidence interval (CI) is a range of values likely to contain the population parameter.

  - Example: A 95% CI means there's a 95% chance the interval contains the true mean.

  - Helps in decision-making along with hypothesis testing.

15. What is the margin of error, and how does it affect the confidence interval?
  - Margin of Error (MoE): The range above and below the sample estimate.

  - CI = estimate ± MoE.

  - Larger MoE → wider CI → less precision.

16. How is Bayes' Theorem used in statistics, and what is its significance?
  - Bayes' Theorem updates the probability of a hypothesis based on new evidence:
    
      P(H|E)= P(E|H)⋅P(H)/P(E)

  - Used in:

      - Bayesian inference

      - Medical diagnosis

      - Machine learning

17. What is the Chi-square distribution, and when is it used?
  - A distribution used for categorical data and variance tests.

  - Common applications:

      - Goodness of fit

      - Test of independence

      - Test of homogeneity

18. What is the Chi-square goodness of fit test, and how is it applied?
  - Tests if a sample matches a population distribution.

  - Compares observed vs. expected frequencies:

    χ 2 =∑(O−E) 2/ E


19. What is the F-distribution, and when is it used in hypothesis testing?
  - A right-skewed distribution used to compare variances.

  - Used in ANOVA and the F-test.

20. What is an ANOVA test, and what are its assumptions?
  - ANOVA (Analysis of Variance) compares means across 3+ groups.

  - Assumptions:

      - Independent observations

      - Normally distributed groups

      - Equal variances

21. What are the different types of ANOVA tests?
  - One-way ANOVA: One independent variable.

  - Two-way ANOVA: Two independent variables.

  - Repeated measures ANOVA: Same subjects tested multiple times.

22. What is the F-test, and how does it relate to hypothesis testing?
  - Used to compare two variances:
      
      F = S12/S22

  - If F is significantly different from 1, it indicates unequal variances.

  - Central to ANOVA and regression model comparisons.

```python
# Practical Part - 2

# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

from scipy.stats import norm
import numpy as np

sample = np.random.normal(loc=52, scale=5, size=30)
pop_mean = 50
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)

z = (sample_mean - pop_mean) / (sample_std / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z-statistic:", z)
print("P-value:", p_value)
```

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

def one_sample_z_test(sample, pop_mean):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    z = (sample_mean - pop_mean) / (sample_std / np.sqrt(len(sample)))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

sample = np.random.normal(52, 4, 50)
z, p = one_sample_z_test(sample, 50)
print("Z-score:", z)
print("P-value:", p)
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

def plot_type_errors(mu_0, mu_a, sigma, n, alpha=0.05):
    se = sigma / np.sqrt(n)
    z_critical = norm.ppf(1 - alpha / 2)
    x = np.linspace(mu_0 - 4*se, mu_0 + 4*se, 1000)

    y0 = norm.pdf(x, mu_0, se)
    ya = norm.pdf(x, mu_a, se)

    plt.plot(x, y0, label="Null Hypothesis", color="blue")
    plt.plot(x, ya, label="Alternative Hypothesis", color="green")
    plt.axvline(mu_0 + z_critical * se, color="red", linestyle="--", label="Critical Values")
    plt.axvline(mu_0 - z_critical * se, color="red", linestyle="--")

    plt.fill_between(x, y0, where=(x < mu_0 - z_critical * se) | (x > mu_0 + z_critical * se), color='red', alpha=0.3, label="Type I Error")
    plt.fill_between(x, ya, where=(x > mu_0 - z_critical * se) & (x < mu_0 + z_critical * se), color='orange', alpha=0.3, label="Type II Error")

    plt.title("Type I and Type II Errors")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_type_errors(mu_0=50, mu_a=52, sigma=5, n=30)
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = sem * t.ppf((1 + confidence) / 2., len(data)-1)
    return mean - margin, mean + margin

data = np.random.normal(60, 5, 50)
ci = confidence_interval(data)
print("Confidence Interval:", ci)
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

def margin_of_error(data, confidence=0.95):
    sem = stats.sem(data)
    moe = sem * t.ppf((1 + confidence) / 2., len(data)-1)
    return moe

moe = margin_of_error(data)
print("Margin of Error:", moe)
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(P_A, P_B_given_A, P_B):
    return (P_B_given_A * P_A) / P_B

# Example: Disease test
P_D = 0.01  # disease prevalence
P_Pos_given_D = 0.99  # sensitivity
P_Pos_given_not_D = 0.05  # false positive

P_Pos = P_D * P_Pos_given_D + (1 - P_D) * P_Pos_given_not_D

P_D_given_Pos = bayes_theorem(P_D, P_Pos_given_D, P_Pos)
print("P(Disease | Positive Test):", P_D_given_Pos)
```
